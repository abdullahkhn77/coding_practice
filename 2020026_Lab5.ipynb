```python
def preprocess_text(text):
    # Convert to lowercase
    text = text.lower()
    
    # Tokenize the text
    tokens = word_tokenize(text)
    
    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    filtered_tokens = [word for word in tokens if word not in stop_words]
    
    # Lemmatize the tokens
    lemmatizer = WordNetLemmatizer()
    lemmatized_tokens = [lemmatizer.lemmatize(token) for token in filtered_tokens]
    
    # Join the tokens back into a string
    preprocessed_text = ' '.join(lemmatized_tokens)
    
    return preprocessed_text
```

```python
docs = []
for i in range(10):
  docs.append(preprocess_text(reuters.raw(documents[i])))
```

```python
def min_edit_distance(source, target):
    m = len(source) + 1
    n = len(target) + 1
    
    # Create a matrix of zeros with m rows and n columns
    d = [[0 for j in range(n)] for i in range(m)]
    
    # Initialize the first row and first column of the matrix
    for i in range(m):
        d[i][0] = i
    for j in range(n):
        d[0][j] = j
    
    # Calculate the minimum edit distance
    for j in range(1, n):
        for i in range(1, m):
            if source[i-1] == target[j-1]:
                d[i][j] = d[i-1][j-1]
            else:
                substitute = d[i-1][j-1] + 1
                insert = d[i][j-1] + 1
                delete = d[i-1][j] + 1
                d[i][j] = min(substitute, insert, delete)
    
    # Return the minimum edit distance
    return d[m-1][n-1]
```

```python
distances = {}
for i in range(len(docs)):
    for j in range(i+1, len(docs)):
        dist = min_edit_distance(docs[i], docs[j])
        key = (i, j)
        distances[key] = dist
```

2.3

1) First limitation of using minimum edit distance is that it only calculates similarity based on number of operations and word order rather than the context due to which if two words are different in order but similar in context, minimum edit distance would consider them similar

2) The minimum edit distance algorithm has a time complexity of O(n^2), where n is the length of the longer of the two input texts. This means that it may not be practical to use for large texts

```python
import math
from collections import Counter

def tf_idf(documents):
    # Calculate term frequency for each document
    tfs = []
    for doc in documents:
        tf = Counter(doc)
        total = sum(tf.values())
        tfs.append({k: v/total for k, v in tf.items()})
        
    # Calculate inverse document frequency
    idf = {}
    n_docs = len(documents)
    for doc in documents:
        for term in set(doc):
            idf[term] = idf.get(term, 0) + 1
    idf = {k: math.log(n_docs/v) for k, v in idf.items()}
    
    # Calculate tf-idf scores for each document
    tf_idfs = []
    for i, doc in enumerate(documents):
        tf_idf = {k: v*idf[k] for k, v in tfs[i].items()}
        tf_idfs.append(tf_idf)
        
    return tf_idfs
```

```python
import math

def cosine_similarity(tf_idfs, doc1_index, doc2_index):
    doc1_tf_idfs = tf_idfs[doc1_index]
    doc2_tf_idfs = tf_idfs[doc2_index]
    
    # Compute the dot product of the TF-IDF vectors
    dot_product = sum(doc1_tf_idfs.get(term, 0) * doc2_tf_idfs.get(term, 0) for term in set(doc1_tf_idfs) & set(doc2_tf_idfs))
    
    # Compute the magnitude of the TF-IDF vectors
    doc1_mag = math.sqrt(sum(tfidf**2 for tfidf in doc1_tf_idfs.values()))
    doc2_mag = math.sqrt(sum(tfidf**2 for tfidf in doc2_tf_idfs.values()))
    
    # Compute the cosine similarity
    if doc1_mag == 0 or doc2_mag == 0:
        return 0
    else:
        return dot_product / (doc1_mag * doc2_mag)
```

```python
for i in range(len(docs)):
    for j in range(i+1, len(docs)):
        similarity = cosine_similarity(tf_idf(terms), i, j)
        print(f"Similarity between doc {i} and doc {j}: {similarity:.2f}")
```

3.3

Advantages : 
i) The algorithm takes into account the importance of a term in a document by giving more weight to terms that occur frequently in the document.

ii) The algorithm scales the TF-IDF values to be between 0 and 1, which makes them easier to interpret and compare.

iii) The algorithm is relatively efficient and can be used to process large amounts of text data

Limitations : 
i) The TF-IDF algorithm may not work well for short documents since the term frequency for each term may be relatively small, making it difficult to distinguish between documents.

ii) The TF-IDF algorithm only considers the frequency of individual terms and does not capture the contextual similarity between terms or phrases, which may be important for some applications

iii) It is sensitive to the size of corpus
